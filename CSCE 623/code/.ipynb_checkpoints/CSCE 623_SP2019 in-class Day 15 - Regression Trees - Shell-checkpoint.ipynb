```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
```

```python
#build data
def makedata(myseed):
    np.random.seed(myseed)
    xmin = 0
    xmax = 5
    #select features to sample at
    obsCount = 400
    X = np.random.uniform(xmin,xmax,size = (obsCount,2))
    #set up y by regions
    y = np.zeros((obsCount,1))
    R1 = np.logical_and(X[:,0]<=2 , X[:,1]<=2)
    R2 = np.logical_and(X[:,0]<=2 , X[:,1]>2)
    R3 = np.logical_and(X[:,0]<=3 , X[:,0]>2)
    R4 = np.logical_and(X[:,0]>3 , X[:,1]<=3.5)
    R5 = np.logical_and(X[:,0]>3 , X[:,1]>3.5)
    #set up y values for each region
    y[R1,0] = 5
    y[R2,0] = 0
    y[R3,0] = 15
    y[R4,0] = 20
    y[R5,0] = 25
    #add noise to data
    noise = np.random.normal(0,1.5,(obsCount,1))
    y = y + noise
    d = np.hstack((X,y))
    df = pd.DataFrame(data = d, columns = ['X1','X2','Y'],dtype=float)
    return(df)
```

```python
def computeRSSonMean(df):
    #YOUR CODE HERE
    RSS = None #YOU-CODE-IT!
    obs_mean = df.Y.mean()
    y_diff = df.Y - obs_mean
    RSS = np.dot(y_diff.T, y_diff)
    return(RSS)    
```

```python
#split a region into exactly 2 halves along exactly 1 feature
def pickSplit2(df):
    #YOUR CODE HERE
    splitFeature = None #YOU-CODE-IT!
    splitValue = None #YOU-CODE-IT!
    bestRSS = np.inf 
    for feature in df.columns[:len(df.columns)-1]:
        feat_min = df[feature].min()
        feat_max = df[feature].max()
        thresholds = np.linspace(feat_min, feat_max, 50)
        for val in thresholds:
            lowRegion, highRegion = divideFrame(df, feature, val)
            rss_low = computeRSSonMean(lowRegion)
            rss_high = computeRSSonMean(highRegion)
            total_rss = rss_low + rss_high
            if total_rss < bestRSS:
                bestRSS = total_rss
                splitValue = val
                splitFeature = feature
    
    return(splitFeature,splitValue,bestRSS)
```

```python
def pickSplit(df):
    splitFeature = None
    splitValue = None
    splitValueX1 = None
    splitValueX2 = None
    
    df1 = df.sort_values(by='X1')
    df2 = df.sort_values(by='X2')
    x1 = np.array(df1.X1)
    x2 = np.array(df2.X2)
    
    x1firsts = x1[0:-1]
    x1seconds = x1[1:]
    x1mids = (x1firsts + x1seconds)/2
    x2mids = (x2[0:-1]+x2[1:])/2
    
    obsCount = len(x1)
    
    initRSS = computeRSSonMean(df)
    bestRSSx1 = initRSS
    bestRSSx2 = initRSS
    bestIndexX1 = 0
    bestIndexX2 = 0
    for dpindex in range(0,len(x1mids)):
        currx1 = x1mids[dpindex]
        currx2 = x2mids[dpindex]
        x1mask = x1<=currx1
        x2mask = x2<=currx2
        dfX1low = df[x1mask]
        dfX1high = df[~x1mask]
        X1RSS = computeRSSonMean(dfX1low)+computeRSSonMean(dfX1high)
        dfX2low = df[x2mask]
        dfX2high = df[~x2mask]
        X2RSS = computeRSSonMean(dfX2low)+computeRSSonMean(dfX2high)
        if X1RSS < bestRSSx1:
            bestRSSx1 = X1RSS
            bestIndexX1 = X1RSS
            splitValueX1 = currx1
        if X2RSS < bestRSSx2:
            bestRSSx2 = X2RSS
            bestIndexX2 = X2RSS
            splitValueX2 = currx2
    if bestRSSx2 <= bestRSSx1:
        splitFeature = 'X2'
        splitValue = splitValueX2
        bestRSS = bestRSSx2
    else:
        splitFeature = 'X1'
        splitValue = splitValueX1
        bestRSS = bestRSSx1
    return(splitFeature,splitValue,bestRSS)      
```

```python
def divideFrame(df,splitFeature,splitVal):
    #YOUR CODE HERE
    
    dfMask = df[splitFeature] <= splitVal
    dfLow = df[dfMask]
    dfHigh = df[~dfMask]
    return(dfLow,dfHigh)
```

```python
#manage the tree

df = makedata(1)

finishedCode=True  #make this True once code is done
if finishedCode:
    feat,val,bestRSS = pickSplit(df)
    print(feat,val,bestRSS)

    dfLow,dfHigh = divideFrame(df,feat,val)
    #decide which region to split next
    fLow,vLow,RSSLow = pickSplit(dfLow)
    fHigh,vHigh,RSSHigh = pickSplit(dfHigh)
    if RSSLow<=RSSHigh:
        feat2,val2,bestRSS2 = pickSplit(dfLow)
        fullRSS2 = bestRSS2+RSSHigh
    else:
        feat2,val2,bestRSS2 = pickSplit(dfHigh)
        fullRSS2 = bestRSS2+RSSLow
    print(feat2,val2,fullRSS2)


threedee = plt.figure().gca(projection='3d')
threedee.scatter(df.X1, df.X2, df.Y)
threedee.set_xlabel('X1')
threedee.set_ylabel('X2')
threedee.set_zlabel('Y')
plt.show()

#feat,val,bestRSS = pickSplit(df)
#dfLow,dfHigh = divideFrame(df,feat,val)
```
